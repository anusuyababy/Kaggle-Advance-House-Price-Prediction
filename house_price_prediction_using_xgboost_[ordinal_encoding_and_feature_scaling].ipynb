## COMBINING BOTH TRAIN AND TEST DATA

```python
train.head()
```

```python
test.head()
```

```python
train.shape, test.shape
```

```python
final = pd.concat([train, test], axis=0)
final.head()
```

```python
feat_cat=[feature for feature in final.columns if final[feature].dtype =='O']
feat_cat
```

```python
for feature in feat_cat:
    print('The feature is {} and number of categories are {}'.format(feature,len(final[feature].unique())))
```

```python
for feature in feat_cat:
    labels_ordered=final[feature].value_counts().to_dict()
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    final[feature]=final[feature].map(labels_ordered)
final.head()
```

```python
final = final.drop(['MSSubClass', 'MSZoning', 'Utilities', 'Street', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'Exterior1st,' 'Exterior2nd', 'ExterCond', 'Foundation', 'BsmtCond',	'BsmtFinType1', 'BsmtFinType2', 'BsmtFinSF2', 'Heating', 'HeatingQC',	'CentralAir', 'Electrical', 'LowQualFinSF', 'BsmtHalfBath', 'KitchenAbvGr', 'Functional', 'GarageType', 'GarageQual', 'PavedDrive', 'EnclosedPorch', 'MiscVal', 'YrSold'], axis = 1)
final.head()
```

```python
df=final.loc[:,final.columns!='SalePrice']
df['SalePrice']=final['SalePrice']
df.head()
```
